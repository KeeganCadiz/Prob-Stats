```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,20) # 20 values from -10 to 10
```

```python
x
```

```python
m = 50
b = 17
```

```python
y = m * x + b
```

```python
plt.plot(x,y,'.') # basic scatter plot with 20 dots
```

```python
r = 100*np.random.randn(20,) # generates 20 numbers from a normal distribution x 100
```

```python
r
```

```python
plt.plot(r,'.') # random scatter plot with a normal distribution around 0
```

```python
y2 = m * x + b + r # adds 'noise' to the basic y = mx + b equation
```

```python
plt.plot(x,y2,'.')
```

```python
m_guess = 1
b_guess = 1

y_guess = m_guess * x + b_guess

plt.plot(x,y_guess,'r.'); # not a very good guess
plt.plot(x,y2,'g.');
```

```python
for m_guess in np.linspace(-100,100,100):

        y_guess = m_guess * x

        plt.plot(x,y_guess,'r.');
        plt.plot(x,y2,'g.');
        plt.show()

        loss = np.sum((y_guess - y)**2) # loss is the difference squared. higher = worse guess

        print(loss,'M = ',m_guess)
```

```python
best_loss = 100000000000
best_m = 0
best_b = 0

for m_guess in np.linspace(-100,100,100):
    for b_guess in np.linspace(-100,100,100):

        y_guess = m_guess * x + b_guess

        loss = np.sum((y_guess - y)**2)

        if loss < best_loss:

            best_loss = loss
            best_m = m_guess
            best_b = b_guess  # keeps track of the combination with lowest loss that represents the best fit to the data
```

```python
y_guess = best_m * x + best_b

plt.plot(x,y_guess,'r.'); # best fit line in red
plt.plot(x,y2,'g.'); # original noisey data
plt.show()

print(best_loss,best_m,best_b)
```

If the universe has been expanding since its beginning at a constant speed, the universe's age would simply be 1/Ho.
Find the inverse of your value of Ho.
Multiply the inverse by $3.09 \times 10^{19}$ km/Mpc to cancel the distance units.
Since you now have the age of the Universe in seconds, divide this number by the number of seconds in a year: $3.16 \times 10^{7}$ sec/yr.

$3.09 \times 10^{19}$ km/Mpc

$3.16 \times 10^{7}$ sec/year

```python
(1/H) * (3.09*10**19)
```

```python
(1/H) * (3.09*10**19)  / (3.16*10**7)
```

```python
 (1/H) * (3.09*10**19)  / (3.16*10**7) / (1000000000)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
def get_database(url,sheet_name): # takes data from the google spreadsheet
    url_head = "https://docs.google.com/spreadsheets/d/"
    url_foot = "/gviz/tq?tqx=out:csv&sheet="
    url_body = url.split('/')[5]
    # sheet_name = "1"
    url_csv = url_head + url_body + url_foot + sheet_name
    df = pd.read_csv(url_csv, sep=',', skiprows=0)
    return df
```
